```python
for pp in range(1,4):
    link = 'https://www.flipkart.com/search?q=microwave+oven&sid=j9e%2Cm38%2Co49&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_9_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_9_na_na_na&as-pos=1&as-type=RECENT&suggestionId=microwave+oven%7CMicrowave+Ovens&requestId=4716036c-57c0-43b2-b240-710b58dc0732&as-searchtext=microwave&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3DMax&p%5B%5D=facets.brand%255B%255D%3DLG&p%5B%5D=facets.brand%255B%255D%3DIFB&p%5B%5D=facets.brand%255B%255D%3DWhirlpool&p%5B%5D=facets.brand%255B%255D%3DGodrej/'
```

```python
page=requests.get(link, str(pp) + '/')
```

```python
page
```

```python
soup = bs(page.content,'html.parser')
soup
```

```python
m_name=[]
m_price=[]
m_dis=[]
m_rate=[]

m_power=[]
m_capacity=[]
m_powerlevels=[]
m_control=[]

count=0
```

```python
import bs4
import numpy as np
```

```python
for i in range (1,4):
    for containers in soup.findAll('a',class_='_1fQZEK'):
            name=containers.find('div',class_='_4rR01T')    
            price=containers.find('div',class_='_30jeq3 _1_WHN1')
            discount=containers.find('div', class_='_3Ay6Sb') 
            rating=containers.find('div', class_='_3LWZlK')         
            
            m_name.append(name.text) # Add product name to list
            m_price.append(price.text)
            m_dis.append(discount.text) if type(discount) == bs4.element.Tag  else m_dis.append(np.nan)
            m_rate.append(rating.text) if type(rating) == bs4.element.Tag  else m_rate.append(np.nan)
        
            specification = containers.find('div',class_='fMghEO')
            mpower=""
            mpl=""
            for col in specification:
                col1=col.find_all('li', class_='rgWa7D')
                for cl in col1:
                    if 'Power:' in cl.text:
                        mpower=cl.text
                    
                    if 'Capacity:' in cl.text:
                        mcapacity=cl.text
                    
                    if 'Power Levels:' in cl.text:
                        mpl=cl.text
                    
                    if 'Control:' in cl.text:
                        mcontrol=cl.text
                    
             
            m_power.append(mpower) if mpower !="" else m_power.append(np.nan)
            m_capacity.append(mcapacity) if mcapacity !="" else m_capacity.append(np.nan)
            m_powerlevels.append(mpl) if mpl !="" else m_powerlevels.append(np.nan)
            m_control.append(mcontrol) if mcontrol !="" else m_control.append(np.nan)

        
            count=count+1 # Increment row count
```

```python
df=pd.DataFrame({'Product Name':m_name,'Price':m_price,'Discount':m_dis,'Rating':m_rate,'Power':m_power,'Capacity':m_capacity,'Power Levels':m_powerlevels,'Control':m_control})
```

```python
df
```

```python
df["Power"]= df["Power"].str.split("|").str[0]
df["Capacity"]= df["Capacity"].str.split("|").str[1]
df["Power Levels"]= df["Power Levels"].str.split("|").str[0]
df["Control"]= df["Control"].str.split("(").str[0]
df
```

```python
df['Power'] = df['Power'].replace({'Power: ' : ''}, regex=True)
df['Capacity'] = df['Capacity'].replace({'Capacity: ' : ''}, regex=True)
df['Power Levels'] = df['Power Levels'].replace({'Power Levels: ' : ''}, regex=True)
df['Control'] = df['Control'].replace({'Control: ' : ''}, regex=True)
df
```

```python
df['Price']=df['Price'].replace({'₹':""},regex=True)
df['Price'] = df['Price'].replace({',' : ''}, regex=True)
df['Discount']=df['Discount'].str.split('%').str[0]
df['Power']=df['Power'].str.split('W').str[0]
df['Capacity']=df['Capacity'].str.split('L').str[0]
df
```

```python
df.info()
```

```python
df['Power Levels'] = df['Power Levels'].replace(np.nan, 0)
```

```python
df
```

```python
df.info()
```

```python
m=df['Power'].mode()
m[0]
df['Power']=df['Power'].replace(np.nan, m[0])
m1=df['Capacity'].mode()
m1[0]
df['Capacity']=df['Capacity'].replace(np.nan,m1[0])
df
```

```python
df['Price']=df['Price'].astype(int)
df['Rating']=df['Rating'].astype(float)
df['Discount']=df['Discount'].astype(int)
df['Power']=df['Power'].astype(int)
df['Capacity']=df['Capacity'].astype(int)
df['Power Levels']=df['Power Levels'].astype(int)
```

```python
m=df['Rating'].mode()
m[0]
df['Rating']=df['Rating'].fillna(m[0])
df
```

```python
df['Control'] = df['Control'].replace({'Jog Dial' : 'Mechanical Knob'}, regex=True)
```

```python
Ctrl = pd.get_dummies(df['Control'])
Ctrl.head()
```

```python
df = pd.concat([df, Ctrl], axis = 1)
df
```
